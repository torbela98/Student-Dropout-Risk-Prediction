#Slide 2: Problem and Objectives



Student attrition affects graduation rates, institutional performance, and the efficient use of educational resources.

### Project Objective
Develop a machine learning model that identifies students who are at risk of dropping out before completing their academic program.

### Data Science Task
- Supervised binary classification
- Dropout = 1
- Graduate = 0
- Students classified as Enrolled were excluded

### Evaluation Strategy
- Primary metric: AUC-ROC
- Target AUC: at least 0.80
- Secondary metric: Recall
- Additional metrics: Accuracy, Precision, and F1-score

#Slide 3: Dataset Overview



- Original records: 4,424 students
- Variables: 36 predictors and 1 target
- Original target classes:
  - Graduate: 2,209
  - Dropout: 1,421
  - Enrolled: 794
- No missing values
- No duplicate records
- Feature groups:
  - Demographic
  - Academic
  - Financial
  - Family background
  - Macroeconomic

#Slide 10: Fairness, Limitations and Next Steps

### Fairness Assessment

- Recall was similar across the two gender groups:
  - Gender 0: 90.0%
  - Gender 1: 89.0%
- Predicted-dropout rates differed:
  - Gender 0: 28.77%
  - Gender 1: 51.74%
- Disparate impact ratio: 0.556

The similar recall values indicate comparable identification of actual dropout cases, but the difference in predicted-dropout rates and the disparate-impact result require further investigation.



### Limitations

- Data came from a specific higher education context
- External generalizability has not yet been established
- Historical patterns may change over time
- Demographic features may capture indirect institutional or socioeconomic effects

### Next Steps

- Validate the model using data from other institutions
- Investigate the gender prediction-rate disparity
- Evaluate reweighting or threshold adjustments
- Monitor performance and fairness after deployment
- Retrain periodically using updated student records
